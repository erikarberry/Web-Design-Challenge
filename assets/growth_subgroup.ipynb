```python
# Dependencies
import pandas as pd
```

```python
# Load Files
acad_outcomes_data_to_load = "acad_outcomes.csv"
growth_data_to_load = "GroupByDistrictNameGrowth.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
acad_outcomes_data = pd.read_csv(acad_outcomes_data_to_load, delimiter=',', encoding="utf-8-sig", dtype={'ACT Percent': int})
#acad_outcomes_data = pd.read_csv(acad_outcomes_data_to_load)
student_growth_data = pd.read_csv(growth_data_to_load)
```

```python
student_growth_data.head()
```

```python
LEA_student_growth_data = pd.DataFrame(student_growth_data.groupby(['District_Name', 'Subgroup']).mean()['School Growth Index Score'])
LEA_student_growth_data.head()
```

```python
table_LEA_student_growth_data = LEA_student_growth_data.unstack(level = -1)
table_LEA_student_growth_data.columns = table_LEA_student_growth_data.columns.get_level_values(1)
table_LEA_student_growth_data.head()
```

```python
# Renames columns to be more readable
table_LEA_student_growth_data = table_LEA_student_growth_data.rename(columns={"Academically or Intellectually Gifted":"Growth_AIG",
                                           "All Sttudents":"Growth_All", 
                                           "American Indian": "Growth_American Indian",
                                           "Asian": "Growth_Asian", 
                                           "Black": "Growth_Black",
                                           "Economically Disadvantaged": "Growth_ED",
                                        "English Learner": "Growth_ELL",
                                           "Hispanic": "Growth_Hispanic",
                                           "Students With Disabilities": "Growth_SwD",
                                           "Two or More Races": "Growth_MultiRacial",
                                        "White": "Growth_White"})
# Iterates through all column data to find NaN values and adds them to a list
table_LEA_student_growth_data.columns[table_LEA_student_growth_data.isna().any()].tolist()
table_LEA_student_growth_data.head()
```

```python
final_student_growth = table_LEA_student_growth_data.reset_index()
final_student_growth.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
final_student_growth.to_csv("/Users/erikaberry/Desktop/Web-Design-Challenge/subgroup_growth_byLEA.csv", index=False, header=True)
```
